```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

suji = pd.read_csv("suji.csv")

start_date = '2003-12-01'
end_date = '2022-12-01'
data = yf.download(['USDKRW=X'],start=start_date, end=end_date)
```

```python
data2 = data.resample(rule="M").mean()
data3 = data2[:-2]
data4 = data3["Close"]
```
